```python
import lightkurve as lk

# Pi Mensae — a bright, nearby star with a known TESS planet
search = lk.search_lightcurve("Pi Mensae", mission="TESS", author="SPOC", sector=1)
lc = search.download().remove_nans().normalize()
lc.plot()
```
